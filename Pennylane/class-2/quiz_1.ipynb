```python
import pennylane as qml
from pennylane import numpy as np

dev = qml.device("default.qubit", wires=["A", "B"])
@qml.qnode(dev)

def circuit(state):
    qml.Hadamard(wires="A")
    qml.CNOT(wires=["A", "B"])
    
    if(state[1] == 1):
        qml.X(wires="A")
    if(state[0] == 1):
        qml.Z(wires="A")
        
    qml.CNOT(wires=["A", "B"])
    qml.Hadamard(wires="A")
    
    return qml.state()

state = np.array([1, 0])
print(circuit(state))
_ = qml.draw_mpl(circuit, style="pennylane")(state)
```
